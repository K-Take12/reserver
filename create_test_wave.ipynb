```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ステップ数と周期設定
steps = 100000
period = 5

# Sin波生成
sine_wave = np.sin(2 * np.pi * np.arange(steps) / period)
pd.DataFrame(sine_wave, columns=["sine_wave"]).to_csv("sine_wave.csv", index=False)

# のこぎり波生成
sawtooth_wave = (np.arange(steps) % period) / (period - 1)
pd.DataFrame(sawtooth_wave, columns=["sawtooth_wave"]).to_csv("sawtooth_wave.csv", index=False)

# 矩形波生成 (ランダムなタイミングで切り替え)
square_wave = np.zeros(steps)
current_value = 0  # 初期値
for i in range(steps):
    # ランダムに切り替え (一定確率で 0 -> 1 または 1 -> 0)
    if np.random.rand() < 0.2:  # 20%の確率で切り替え
        current_value = 1 - current_value
    square_wave[i] = current_value
pd.DataFrame(square_wave, columns=["square_wave"]).to_csv("square_wave.csv", index=False)

```

```python
# 波形のプロット (先頭500ステップを表示)
plt.figure(figsize=(12, 8))

# Sin波プロット
plt.subplot(3, 1, 1)
plt.plot(sine_wave[:500], label="Sine Wave")
plt.title("Sine Wave (First 500 steps)")
plt.xlabel("Step")
plt.ylabel("Amplitude")
plt.grid()
plt.legend()

# のこぎり波プロット
plt.subplot(3, 1, 2)
plt.plot(sawtooth_wave[:500], label="Sawtooth Wave", color="orange")
plt.title("Sawtooth Wave (First 500 steps)")
plt.xlabel("Step")
plt.ylabel("Amplitude")
plt.grid()
plt.legend()

# 矩形波プロット
plt.subplot(3, 1, 3)
plt.plot(square_wave[:50], label="Square Wave", color="green")
plt.title("Square Wave (First 500 steps)")
plt.xlabel("Step")
plt.ylabel("Amplitude")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()

print("CSVファイルを作成しました: sine_wave.csv, sawtooth_wave.csv, square_wave.csv")
```
